```python
from pandas_datareader import data as web
import datetime as dt
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd 
import numpy as np 
yf.pdr_override()
```

```python
#設定資料起始與結束日期
start = dt.datetime(2019, 1, 1) 
end = dt.datetime(2021, 3, 30) 
```

```python
#df為SPY，df2為0050，拉資料下來
df = web.get_data_yahoo(['SPY'],start, end) 
df2=web.get_data_yahoo(['0050.TW'],start, end)
```

```python
#資料指定為DataFrame方便運算
d = pd.DataFrame(df) 
d2 = pd.DataFrame(df2) 
#印出來看看
print(d) 
print(d2) 
```

```python
#取adj_close作比較
adj_df=df["Adj Close"] 
adj_df2=df2["Adj Close"] 
```

```python
#畫圖，google搜尋matplotlib
fig,ax1 = plt.subplots(nrows=1,ncols=1,figsize=(15,6))
ax1.set_ylabel('SPY', color='tab:blue')
ax1.plot(adj_df, color='tab:blue', alpha=0.75)
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax2 = ax1.twinx()
ax2.set_ylabel('0050', color='tab:blue')
ax2.plot( adj_df2, color='black', alpha=1)
ax2.tick_params(axis='y', labelcolor='black')
plt.show()
```

```python
dfcross=pd.DataFrame()
dfcross["SPY"]=adj_df
dfcross["0050"]=adj_df2
#計算相關係數
dfcross.corr(method='pearson', min_periods=1)
```
